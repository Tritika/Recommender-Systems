```python
# Filtering records with only positive values
df = df.loc[df['Quantity'] > 0]
```

```python
df.loc[df['CustomerID'].isna()].head()
```

```python
# Dropping NAN values
df = df.dropna(subset=['CustomerID'])
```

```python
df.shape
```

# User-based collaborative filtering

```python
# Building a matrix of items for the client
customer_item_matrix = df.pivot_table(
index = 'CustomerID',
columns = 'StockCode',
values = 'Quantity',
aggfunc = 'sum'
)

customer_item_matrix.head()
```

```python
# Encoding above data to 0-1
customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x>0 else 0)

customer_item_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Measuring similiarity between users

user_user_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix))
```

```python
# Renaming index and columns of above dataframe

user_user_sim_matrix.columns = customer_item_matrix.index

user_user_sim_matrix.head()
```

```python
user_user_sim_matrix['CustomerID'] = customer_item_matrix.index
user_user_sim_matrix = user_user_sim_matrix.set_index('CustomerID')

user_user_sim_matrix.head(10)
```

```python
# Classifying customers most similars with ID 12350

user_user_sim_matrix.loc[12350.0].sort_values(ascending = False).head()
```

```python
items_bought_by_A = set(customer_item_matrix.loc[12350.0].iloc[customer_item_matrix.loc[12350.0].to_numpy().nonzero()].index)

items_bought_by_A
```

```python
items_bought_by_B = set(customer_item_matrix.loc[17935.0].iloc[customer_item_matrix.loc[17935.0].to_numpy().nonzero()].index)

items_bought_by_B
```

```python
items_to_recommend_to_B = items_bought_by_A - items_bought_by_B

items_to_recommend_to_B
```

```python
# Getting descriptions of items to recommend to B
df.loc[df['StockCode'].isin(items_to_recommend_to_B), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode')
```

# Item-based collaborative filtering

```python
item_item_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix.T))

item_item_sim_matrix.head(10)
```

```python
# Naming the indexes and columns with product codes using metrics.pairwise from sklearn 

item_item_sim_matrix.columns = customer_item_matrix.T.index

item_item_sim_matrix['StockCode'] = customer_item_matrix.T.index
item_item_sim_matrix = item_item_sim_matrix.set_index('StockCode')
```
